```python
import os
from selenium import webdriver                                                             
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
https://pythonspot.com/selenium-get-images/

skin_links = []

driver=webdriver.Chrome(ChromeDriverManager().install())

driver.maximize_window()

driver.get('https://lolskinshop.com/product-category/lol-skins/')

skins_links = []

for page_num in range (0, 81):
    
    current_page_skins = driver.find_element(By.CLASS_NAME, "skins-divisions__content")

    images = current_page_skins.find_elements(By.TAG_NAME, "img")

    for image in images:
        skins_links.append(image.get_attribute("src"))

    if (page_num != 80): 
        next_button = driver.find_element(By.CLASS_NAME,"slick-next")

        driver.execute_script("arguments[0].click();", next_button)

driver.quit()
```

```python
print(skins_links)
```

```python
output_file = open('all_skin_links.txt', 'w')

for skin_link in skins_links:
    output_file.write(skin_link + '\n')

output_file.close()
```
